```python
# Define a singe table structure in raw SQL
import sqlite3
from sqlalchemy import create_engine, text # Import text
import pandas as pd

# Connect to the SQLite database
# Use the same database file path
engine = create_engine('sqlite:///Lab 4 Site Investigation.db') # Using the same engine as before
conn = engine.connect()

Create_table_query = """CREATE TABLE IF NOT EXISTS CPT_readings (
    Layer TEXT,
    Depth_m REAL,
    Depth_ft REAL,
    qc_tsf REAL,
    qt_tsf REAL,
    fs_tsf REAL,
    u_ft REAL,
    Rf_pct)"""

# Read data from SQL tables into DataFrames
cpt_cook_01 = pd.read_sql_query("SELECT * FROM cpt_cook_01", engine)
spt_cook_02 = pd.read_sql_query("SELECT * FROM spt_cook_02", engine)
cpt_cook_03 = pd.read_sql_query("SELECT * FROM cpt_cook_03", engine)
cpt_cook_04 = pd.read_sql_query("SELECT * FROM cpt_cook_04", engine)
cpt_cook_06 = pd.read_sql_query("SELECT * FROM cpt_cook_06", engine)
spt_cook_07 = pd.read_sql_query("SELECT * FROM spt_cook_07", engine)
spt_cook_02_off01 = pd.read_sql_query("SELECT * FROM spt_cook_02_off01", engine)
spt_cook_02_off02 = pd.read_sql_query("SELECT * FROM spt_cook_02_off02", engine)
cpt_cook_06_off01 = pd.read_sql_query("SELECT * FROM cpt_cook_06_off01", engine)

# Combine CPT excel files for CPT_readings Table
combined_data = pd.concat([
    cpt_cook_01,
    spt_cook_02,
    cpt_cook_03,
    cpt_cook_04,
    cpt_cook_06,
    spt_cook_07,
    spt_cook_02_off01,
    spt_cook_02_off02,
    cpt_cook_06_off01
])
print(combined_data)

# Execute the query using the connection object
with engine.begin() as connection:
    connection.execute(text(Create_table_query)) # Wrap the query with text()

print("Table created successfully.")
```

```python
# Graphing Pore Pressure using Robertson et al 1986
import matplotlib.pyplot as plt
import numpy as np

# CPT data (depth in ft, cone resistance qc in MPa, sleeve friction fs in kPa)
depth = df['Depth_ft']
qc = df['qc_tsf']
fs = df['fs_tsf']

# Robertson et al. (1986) method to estimate pore pressure
# Assuming a simple linear relationship for demonstration purposes
# Pore pressure u = a * qc + b * fs
a = 0.1  # coefficient for cone resistance
b = 0.05  # coefficient for sleeve friction

# Calculate pore pressure
pore_pressure = a * qc + b * fs

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(pore_pressure, depth, label='Pore Pressure', color='b', marker='o')
plt.gca().invert_yaxis()  # Invert y-axis to have depth increasing downwards
plt.xlabel('Pore Pressure (kPa)')
plt.ylabel('Depth (ft)')
plt.title('Pore Pressure Profile using Robertson et al. (1986)')
plt.legend()

# Save to PDF
plt.tight_layout()
plt.savefig("Pore_Pressure_behavior_robertson1986.pdf")

# Show plot
plt.show()
```

```python
# Soil type Behavior using Robertson et al 1990
import numpy as np

def calculate_qt(qc, sigma_vo, sigma_vo_eff):
    return (qc - sigma_vo) / sigma_vo_eff

def calculate_Fs(fs, qt):
    return (fs / qt) * 100

def calculate_Ic(qt, fs):
    return np.sqrt((3.47 - np.log10(qt))**2 + (np.log10(fs) + 1.22)**2)

def classify_soil(Ic):
    if Ic < 1.31:
        return "Gravelly sand to sand"
    elif 1.31 <= Ic < 2.05:
        return "Sands: clean sands to silty sands"
    elif 2.05 <= Ic < 2.60:
        return "Silty sand to sandy silt"
    elif 2.60 <= Ic < 2.95:
        return "Silt mixtures: clayey silt to silty clay"
    elif 2.95 <= Ic < 3.60:
        return "Clays: silty clay to clay"
    else:
        return "Organic soils: peats"

# Assuming 'df' is your DataFrame and 'depth' is a column in it
depth = df['Depth_ft']

# Calculate qt, Fs, and Ic for each row in the DataFrame
# Assuming 'qc_tsf', 'fs_tsf', 'sigma_vo', 'sigma_vo_eff' are columns in 'df'
# If 'sigma_vo' and 'sigma_vo_eff' are not in 'df', you'll need to calculate or define them
df['qt'] = calculate_qt(df['qc_tsf'], df['sigma_vo'] if 'sigma_vo' in df else 0.1, df['sigma_vo_eff'] if 'sigma_vo_eff' in df else 0.08)
df['Fs'] = calculate_Fs(df['fs_tsf'], df['qt'])
df['Ic'] = calculate_Ic(df['qt'], df['Fs'])
df['soil_type'] = df['Ic'].apply(classify_soil)

# Now plot Ic vs. depth
plt.figure(figsize=(10, 6))
plt.plot(df['Ic'], depth, label='Soil Behavior Type Ic', color='b', marker='o')
plt.gca().invert_yaxis()
plt.xlabel('Soil Behavior Type Index (Ic)')
plt.ylabel('Depth (ft)')
plt.title('Soil Behavior Type (Robertson et al., 1990)')
plt.legend()
plt.grid(True)

first_soil_type = df['soil_type'].iloc[0]
second_soil_type = df['soil_type'].iloc[1]
third_soil_type = df['soil_type'].iloc[2]

print(f"Soil Behavior Type: {soil_type}")

# Save to PDF
plt.tight_layout()
plt.savefig("soil_type_behavior_robertson1990.pdf")

# Show plot
plt.show()
```
