```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.cbook import get_sample_data
from matplotlib.colors import LightSource
from matplotlib import colors as mcolors
```

```python
data = pd.read_csv('Ex1_StressStrain-1.csv')
data.head()
```

```python
plt.plot(data.Strain, data.Stress_0deg, 'b--', label='0 deg')
plt.plot(data.Strain, data.Stress_100deg, 'r--', label='100 deg')
plt.plot(data.Strain, data.Stress_200deg, 'g--', label='200 deg')
plt.plot(data.Strain, data.Stress_300deg, 'p--', label='300 deg')
plt.plot(data.Strain, data.Stress_400deg, 'r.-', label='400 deg')
plt.plot(data.Strain, data.Stress_500deg, 'm--', label='500 deg')
plt.plot(data.Strain, data.Stress_600deg, 'g.-', label='600 deg')
plt.plot(data.Strain, data.Stress_700deg, 'g--', label='700 deg')
plt.plot(data.Strain, data.Stress_800deg, 'b.-', label='800 deg')
plt.plot(data.Strain, data.Stress_900deg, 'r.-', label='900 deg')
plt.xlabel('Axial Strain(%)')
plt.ylabel('Differential Stress (MPa)')
plt.legend()
plt.savefig('StressStrainExp.png')
plt.savefig('StressStrainExp.pdf')
plt.show()
```

```python
E_0 = (data.Stress_0deg[8]-data.Stress_0deg[7])/(data.Strain[8]-data.Strain[7])
E_100 = (data.Stress_100deg[8]-data.Stress_100deg[7])/(data.Strain[8]-data.Strain[7])
E_200 = (data.Stress_200deg[8]-data.Stress_200deg[7])/(data.Strain[8]-data.Strain[7])
E_300 = (data.Stress_300deg[8]-data.Stress_300deg[7])/(data.Strain[8]-data.Strain[7])
E_400 = (data.Stress_400deg[8]-data.Stress_400deg[7])/(data.Strain[8]-data.Strain[7])
E_500 = (data.Stress_500deg[8]-data.Stress_500deg[7])/(data.Strain[8]-data.Strain[7])
E_600 = (data.Stress_600deg[8]-data.Stress_600deg[7])/(data.Strain[8]-data.Strain[7])
E_700 = (data.Stress_700deg[8]-data.Stress_700deg[7])/(data.Strain[8]-data.Strain[7])
E_800 = (data.Stress_800deg[8]-data.Stress_800deg[7])/(data.Strain[8]-data.Strain[7])
E_900 = (data.Stress_900deg[8]-data.Stress_900deg[7])/(data.Strain[8]-data.Strain[7])

Temp = np.array([0, 100, 200, 300, 400, 500, 600, 700, 800, 900])
E = np.array([E_0, E_100, E_200, E_300, E_400, E_500, E_600, E_700, E_800, E_900])/0.01
print(Temp)
print(E)

E_GPa = E/1000
print(E_GPa)
```

```python
plt.plot(Temp, E_GPa, 'go-', label = 'Youngs Modulus')
plt.xlabel('Temperature (degC)')
plt.ylabel('Youngs Modulus (GPa)')
plt.legend()
plt.savefig('TempExp.png')
plt.savefig('TempExp.pdf')
```
